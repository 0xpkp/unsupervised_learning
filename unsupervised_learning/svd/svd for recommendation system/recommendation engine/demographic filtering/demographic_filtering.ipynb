## another method using weighted rating : 
    weighted average = ((r*v) + (c*m)) / (v+m)
        where, v = total number of reviews for each movie
               r = average rating for each movie
               c = mean of 'r'
               m = minimum number of votes required

```python
asd = df.copy()
```

```python
asd.head()
```

```python
movie_id = asd.movie_id.value_counts().index
```

```python
total_ratingcount = []
averagerating = []
for i in movie_id:
    total_ratingcount.append(asd[asd.movie_id == i].rating.shape[0])
    averagerating.append(asd[asd.movie_id == i].rating.mean())
```

```python
qwe = pd.DataFrame()

qwe['movie_id'] = movie_id
qwe['total_rating'] = total_ratingcount
qwe['average_rating'] = averagerating
```

```python
qwe.head()
```

```python
v = qwe['total_rating']
r = qwe['average_rating']
c = qwe['average_rating'].mean()
m = qwe['total_rating'].quantile(0.7)

qwe['weighted_average'] = ((r*v) + (c*m)) / (v+m)
```

```python
qwe=qwe.sort_values(by = 'weighted_average', ascending=False)
```
