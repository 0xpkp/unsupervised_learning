```python
import pandas as pd
import numpy as np
```

```python
# rating dataset - contians user_is, item_id, rating and timestamp. where 'item_id' is movie id.

column_names = ['user_id', 'item_id', 'rating', 'timestamp']
data = pd.read_csv(r"D:\python\ai\unsupervised learning\singular value decomposition(svd)\Movie-Recommender-in-python-master krish naik\u.data", sep='\t', names = column_names)
data.head()
```

```python
# movie title dataset - contians item_id corresponding to its movie name.

movie_titles = pd.read_csv(r"D:\python\ai\unsupervised learning\singular value decomposition(svd)\Movie-Recommender-in-python-master krish naik\Movie_Id_Titles")
movie_titles.head()
```

```python
# merging movie_title to data

data= data.merge(movie_titles, on= 'item_id')
data.head()
```

```python
total_ratings = data.groupby('title')['title'].count()
total_ratings = pd.DataFrame(total_ratings)
total_ratings.columns = ['count']
total_ratings = total_ratings[total_ratings['count']>100]    
total_ratings.head() # dropping all the movies that has total reviews less than 100. we should recommend movie only if the movies is in this list.
```

```python
rating_pivot = data.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
rating_pivot
```

```python
# now to select movies that are similar to a movie 'X', we will correlate the ratings of the movie 'X' with all other ratings.
similar_to_starwars = rating_pivot.corrwith(rating_pivot['Star Wars (1977)'])

# filtering movies that have ratings less than 100
similar_to_starwars = pd.DataFrame(similar_to_starwars)
similar_to_starwars.columns = ['correlation']
similar_to_starwars = similar_to_starwars.loc[total_ratings.index]
```

```python
similar_to_starwars.sort_values(by  = 'correlation', ascending = False).head(11) #these are the top 10 movies similar to 'Star Wars (1977)'.
```
